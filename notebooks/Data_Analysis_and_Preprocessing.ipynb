Feature Engineering
Transaction frequency and velocity for Fraud_Data.csv
Time-Based features for Fraud_Data.csv
hour_of _day
day_of_week


```python
# first Create Time-Based Features

df_Fraud['hour_of_day'] = df_Fraud['purchase_time'].dt.hour
df_Fraud['day_of_week'] = df_Fraud['purchase_time'].dt.dayofweek
```

```python
transaction_frequency =df_Fraud.groupby('user_id').size().reset_index(name='transaction_frequency')
fraud_data = df_Fraud.merge(transaction_frequency, on='user_id', how='left')
```

```python
# Count transactions per customer per hour
transaction_velocity = fraud_data.groupby(['user_id', 'hour_of_day']).size().reset_index(name='transaction_velocity')

# Merge velocity back to the original DataFrame
fraud_data = fraud_data.merge(transaction_velocity, on=['user_id', 'hour_of_day'], how='left')
```
